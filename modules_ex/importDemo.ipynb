```python
import os
import sys

# Check what datasets are available
print("üìÅ Available datasets:")
input_path = "/kaggle/input/"
for dataset in os.listdir(input_path):
    print(f"   ‚Ä¢ {dataset}")
    
    # Show the structure of your dataset
    dataset_path = os.path.join(input_path, dataset)
    print(f"     Contents:")
    for item in os.listdir(dataset_path):
        item_path = os.path.join(dataset_path, item)
        if os.path.isdir(item_path):
            print(f"       üìÅ {item}/")
            # Show contents of important folders
            if item in ['modules_ex', 'data', 'nb']:
                for subitem in os.listdir(item_path)[:5]:  # Show first 5 items
                    print(f"          - {subitem}")
                if len(os.listdir(item_path)) > 5:
                    print(f"          ... and {len(os.listdir(item_path)) - 5} more")
        else:
            print(f"       üìÑ {item}")
    print()
```

```python
DATASET_NAME = "edhec-input"

#set up paths
dataset_path = f"/kaggle/input/{DATASET_NAME}"
modules_path = f"{dataset_path}/modules_ex"
data_path = f"{dataset_path}/data"

# Add modules to Python path
if os.path.exists(modules_path):
    sys.path.insert(0, modules_path)
    print(f"‚úÖ Added {modules_path} to Python path")

    # List available modules
    print(f"\nüì¶ Available Python modules:")
    for item in os.listdir(modules_path):
        if item.endswith('.py') and not item.startswith('__'): # omit init files
            module_name = item[:-3]
            print(f" {module_name}")

else:
    print(f"‚ùå modules_ex not found at {modules_path}")

# check data folder
if os.path.exists(data_path):
    print(f"\nüìä Available data files:")
    for item in os.listdir(data_path):
        if item.endswith(('.csv', '.json', '.txt', '.xlsx')):
            print(f" {item}")
else:
    print(f"‚ùå data folder not found at {data_path}")
```
